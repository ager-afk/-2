# Класс Попугай
Реализуем класс `Parrot` с постепенным усложнением функциональности.

```python
import random
```

## Уровень A: Базовая версия
Попугай всегда говорит одну и ту же фразу.

```python
class Parrot:
    def say(self):
        print("Привет, друзья!")

# Проверяем работу
p = Parrot()
p.say()
```

## Уровень B: Фраза при создании
Теперь фразу можно задавать в момент создания попугая.

```python
class Parrot:
    def __init__(self, text):
        self.text = text
    
    def say(self):
        print(self.text)

# Создаём двух разных попугаев
p1 = Parrot("Гав!")
p2 = Parrot("Мяу!")
p1.say()  
p2.say()  
```

## Уровень C: Смена фразы
Добавляем метод `newText` для изменения фразы во время работы.

```python
class Parrot:
    def __init__(self, text):
        self.text = text
    
    def say(self):
        print(self.text)
    
    def newText(self, text):
        self.text = text

# Меняем фразу у одного попугая
p = Parrot("Гав!")
p.say()           
p.newText("Мяу!")
p.say()           
```

## Уровень D: Повторение фраз
Метод `say` теперь умеет повторять фразу несколько раз.

```python
class Parrot:
    def __init__(self, text):
        self.text = text
    
    def say(self, repeat=1):
        for i in range(repeat):
            print(self.text)
    
    def newText(self, text):
        self.text = text

# Проверяем с разным количеством повторений
p = Parrot("Гав!")
p.say()           
p.newText("Мяу!")
p.say(3)          
```

## Уровень E: Набор фраз
Попугай запоминает несколько фраз и выбирает случайную.

```python
import random

class Parrot:
    def __init__(self, text):
        self.phrases = [text]
    
    def say(self, repeat=1):
        phrase = random.choice(self.phrases)
        for i in range(repeat):
            print(phrase)
    
    def learn(self, text):
        self.phrases.append(text)

# Учим новым словам и проверяем
p = Parrot("Гав!")
p.say()              
p.learn("Мяу!")
p.say()              
p.say(3)             
```

#  Класс РядЛампочек
Моделируем гирлянду с разными возможностями.

## Задание A: Восемь лампочек
Базовая модель из 8 лампочек (0 — выкл, 1 — вкл).

```python
class LampRow:
    def __init__(self):
        self.__state = "00000000"
    
    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, value):
        if len(value) == 8 and all(c in "01" for c in value):
            self.__state = value
        else:
            self.__state = "00000000"
    
    def show(self):
        for c in self.__state:
            print("*" if c == "1" else "-", end="")
        print()

# Демонстрация работы
lamps = LampRow()
lamps.show()                    
lamps.state = "10101010"
print(lamps.state)             
lamps.show()                   
```

## Задание B: Произвольная длина
Теперь количество лампочек можно задать в конструкторе.

```python
class LampRow:
    def __init__(self, count=8):
        self.count = count
        self.__state = "0" * count
    
    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, value):
        if len(value) == self.count and all(c in "01" for c in value):
            self.__state = value
        else:
            self.__state = "0" * self.count
    
    def show(self):
        for c in self.__state:
            print("*" if c == "1" else "-", end="")
        print()

# Проверяем с 6 лампочками
lamps = LampRow(6)
lamps.show()                    
lamps.state = "101010"
print(lamps.state)              
lamps.show()                   
lamps.state = "10101010"       
print(lamps.state)             
lamps.show()                    
```

## Задание C: Два цвета
Добавляем второй цвет: 1 — красный (*), 2 — зелёный (o).

```python
class LampRow:
    def __init__(self, count=8):
        self.count = count
        self.__state = "0" * count
    
    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, value):
        if len(value) == self.count and all(c in "012" for c in value):
            self.__state = value
        else:
            self.__state = "0" * self.count
    
    def show(self):
        for c in self.__state:
            if c == "1":
                print("*", end="")
            elif c == "2":
                print("o", end="")
            else:
                print("-", end="")
        print()

# Тестируем цветную гирлянду
lamps = LampRow(6)
lamps.show()                    
lamps.state = "102102"
print(lamps.state)              
lamps.show()                    
lamps.state = "10201010"        
print(lamps.state)              
lamps.show()                    
```

## Задание D: Внутреннее хранение
Внешне всё выглядит так же, но внутри состояние хранится как целое число.

```python
class LampRow:
    def __init__(self, count=8):
        self.count = count
        self.__state = "0" * count
    
    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, value):
        if len(value) == self.count and all(c in "012" for c in value):
            self.__state = value
        else:
            self.__state = "0" * self.count
    
    def show(self):
        for c in self.__state:
            if c == "1":
                print("*", end="")
            elif c == "2":
                print("o", end="")
            else:
                print("-", end="")
        print()

# Проверяем, что интерфейс не изменился
lamps = LampRow(6)
lamps.show()                    
lamps.state = "102102"
print(lamps.state)              
lamps.show()                    
lamps.state = "10201010"        
print(lamps.state)              
lamps.show()                    
```
